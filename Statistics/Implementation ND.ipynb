```python
#Import Libraries
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
#Load the datasets
```

```python
glaxo = pd.read_csv('GLAXO.csv')
glaxo.head()
```

```python
beml = pd.read_csv('BEML.csv')
beml
```

```python
#Select relevant columns
```

```python
glaxo = glaxo[['Date','Close']]
beml = beml[['Date','Close']]
```

```python
glaxo
```

```python
#Data description
```

```python
glaxo.describe()
```

```python
glaxo.info()
```

```python
#Change the datatype of Date column to Datetime 
```

```python
glaxo['Date'] = pd.DatetimeIndex(glaxo['Date'])
```

```python
beml['Date'] = pd.DatetimeIndex(beml['Date'])
```

```python
glaxo
```

```python
glaxo = glaxo.set_index(glaxo['Date'])
```

```python
beml = beml.set_index(beml['Date'])
```

```python
glaxo
```

```python
beml
```

```python
glaxo.drop(columns=['Date'], inplace=True)
```

```python
glaxo
```

```python
glaxo.columns
```

```python
beml.drop(columns=['Date'], inplace=True)
```

```python
beml.columns
```

```python
beml
```

```python
#Plot closing value for both glaxo and beml
sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = 200
plt.plot(glaxo['Close'])
plt.xlabel('Time')
plt.ylabel('Close')
```

```python
plt.plot(beml['Close'])
plt.xlabel('Time')
plt.ylabel('Close')
```

```python
#Calculate gains 
```

```python
glaxo['gain'] = glaxo['Close'].pct_change(periods=1)
beml['gain'] = beml['Close'].pct_change(periods=1)
```

```python
glaxo
```

```python
beml
```

```python
#Drop the null records
```

```python
glaxo.dropna(inplace=True)
beml.dropna(inplace=True)
```

```python
glaxo
```

```python
#Plot gain values
```

```python
plt.plot(glaxo['gain'])
plt.axhline(glaxo['gain'].mean(), color = 'r')
plt.xlabel('Time')
plt.ylabel('Gain')
```

```python
plt.plot(beml['gain'])
plt.axhline(beml['gain'].mean(), color = 'y')
plt.xlabel('Time')
plt.ylabel('Gain')
```

```python
#Subplot for distribution of BEML and Glaxo
```

```python
plt.subplot(1,2,1)
sns.distplot(glaxo['gain'],color = 'r', label = 'GLAXO')
plt.xlabel('Time')
plt.ylabel('Gain')
plt.legend()
plt.title('Distribution Graph')

plt.subplot(1,2,2)
sns.distplot(beml['gain'], color = 'b', label = 'BEML')
plt.legend()
plt.title('Distribution Graph')
```
